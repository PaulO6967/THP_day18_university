# # Import des fichier

```python
import numpy as np
import pandas as pd
saudi_arabiata_ranking = pd.read_csv("cwurData.csv")
```

```python
times_ranking = pd.read_csv("timesData.csv")
```

```python
shanghai_ranking = pd.read_csv("shanghaiData.csv")
```

```python
education_expenditure = pd.read_csv("education_expenditure_supplementary_data.csv")
```

```python
educational_attainement = pd.read_csv("educational_attainment_supplementary_data.csv")
```

```python
university_list_by_country = pd.read_csv("school_and_country_table.csv")
```

```python
print("Classement par l'Arabi saoudite (année) : ", saudi_arabiata_ranking["year"].unique())
print("Classement par le Times (année) : ", times_ranking["year"].unique())
print("Classement par Shangai (année) : ", shanghai_ranking["year"].unique())
```

```python
print("Classement par l'Arabi saoudite (moyenne de d'université pris en compte par an) : ", saudi_arabiata_ranking["year"].value_counts().mean().round(2))

print("Classement par le Times (moyenne de d'université pris en compte par an) : ", times_ranking["year"].value_counts().mean().round(2))

print("Classement par Shangai (moyenne de d'université pris en compte par an) : ",shanghai_ranking["year"].value_counts().mean().round(2))
```

```python
print("Classement par l'Arabi saoudite (position de l'université de strasbourg par an) : ")
saudi_arabiata_ranking_two_colomuns = saudi_arabiata_ranking.iloc[:,[0, 1, -1]]
display(saudi_arabiata_ranking_two_colomuns.loc[saudi_arabiata_ranking_two_colomuns["institution"].str.contains("Strasbourg")])

print("Classement par le Times (position de l'université de strasbourg par an) : ")
times_ranking_two_colomuns = times_ranking.iloc[:,[0, 1, -1]]
display(times_ranking_two_colomuns.loc[times_ranking_two_colomuns["university_name"].str.contains("Strasbourg")])

print("Classement par Shangai (position de l'université de strasbourg par an) ")
shanghai_ranking_two_colomuns = shanghai_ranking.iloc[:,[0, 1, -1]]
display(shanghai_ranking_two_colomuns.loc[shanghai_ranking_two_colomuns["university_name"].str.contains("Strasbourg", na=False)])
```

```python
times_university_2016 = times_ranking.loc[times_ranking["year"] == 2016]
times_university_2016_format = times_university_2016.iloc[:20,:3]

print("Classement par le Times (position des 20 meilleurs université de 2016) : ")
display(times_university_2016_format)
```

```python
print("La proportion d'université amerciane dans le top 10 de 2016 est de ",(times_university_2016_format["country"].value_counts()[0])/times_university_2016_format.shape[0]*100,"%")
```

```python
shanghai_ranking_2015 = shanghai_ranking.loc[shanghai_ranking["year"] == 2015, ["university_name"]]
shanghai_ranking_2015 = shanghai_ranking_2015.rename(columns={"university_name":"school_name"})
result = pd.merge(shanghai_ranking_2015, university_list_by_country, on="school_name", how='left' )

display(result["country"].value_counts())
```

```python
print("il y a ", saudi_arabiata_ranking["country"].value_counts()["France"], "université française, dans le classement")
```

```python
education_expenditure_2011 = education_expenditure.iloc[:,[0,1,2,8]]
education_expenditure_2011 = education_expenditure_2011.loc[education_expenditure_2011["institute_type"] =="Higher Education Institutions"]
education_expenditure_2011 = education_expenditure_2011.loc[education_expenditure_2011["direct_expenditure_type"] == "Total"]

times_ranking_2011 = times_ranking.loc[times_ranking["year"] == 2011]
times_ranking_2011 = times_ranking_2011["country"].value_counts().sort_index()
times_ranking_2011 = times_ranking_2011.rename({"United States of America" : "United States"})
times_ranking_2011 = times_ranking_2011.to_frame()

result2 = education_expenditure_2011.merge(times_ranking_2011, left_on="country", right_index=True, how="left")
result2["ratio"] = result2["2011"] / result2["country_y"] 
display(result2.iloc[:,[0,5]].sort_values(by='ratio'))
```
