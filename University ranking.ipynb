```python
print("Classement par l'Arabi saoudite (année) : ", saudi_arabiata_ranking["year"].unique())
print("Classement par le Times (année) : ", times_ranking["year"].unique())
print("Classement par Shangai (année) : ", shanghai_ranking["year"].unique())
```

```python
print("Classement par l'Arabi saoudite (moyenne de d'université pris en compte par an) : ", saudi_arabiata_ranking["year"].value_counts().mean().round(2))

print("Classement par le Times (moyenne de d'université pris en compte par an) : ", times_ranking["year"].value_counts().mean().round(2))

print("Classement par Shangai (moyenne de d'université pris en compte par an) : ",shanghai_ranking["year"].value_counts().mean().round(2))
```

```python
print("Classement par l'Arabi saoudite (position de l'université de strasbourg par an) : ")
saudi_arabiata_ranking_two_colomuns = saudi_arabiata_ranking.iloc[:,[0, 1, -1]]
display(saudi_arabiata_ranking_two_colomuns.loc[saudi_arabiata_ranking_two_colomuns["institution"].str.contains("Strasbourg")])

print("Classement par le Times (position de l'université de strasbourg par an) : ")
times_ranking_two_colomuns = times_ranking.iloc[:,[0, 1, -1]]
display(times_ranking_two_colomuns.loc[times_ranking_two_colomuns["university_name"].str.contains("Strasbourg")])

print("Classement par Shangai (position de l'université de strasbourg par an) ")
shanghai_ranking_two_colomuns = shanghai_ranking.iloc[:,[0, 1, -1]]
display(shanghai_ranking_two_colomuns.loc[shanghai_ranking_two_colomuns["university_name"].str.contains("Strasbourg", na=False)])
```

```python
times_university_2016 = times_ranking.loc[times_ranking["year"] == 2016]
times_university_2016_format = times_university_2016.iloc[:10,:3]

print("Classement par le Times (position des 10 meilleurs université de 2016) : ")
display(times_university_2016_format)
```

```python
print("La proportion d'université amerciane dans le top 10 de 2016 est de ",(times_university_2016_format["country"].value_counts()[0])/times_university_2016_format.shape[0]*100,"%")
```
